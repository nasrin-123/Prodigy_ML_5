```python
from google.colab import files
import io
import pandas as pd

# Upload the file
uploaded = files.upload()

# Print uploaded file names
print(uploaded.keys())

```

```python


# Access the uploaded file name (use the correct key)
file_name = list(uploaded.keys())[0]

# Load the CSV file into a DataFrame
df = pd.read_csv(io.BytesIO(uploaded[file_name]))

# Print the DataFrame to check its content
print(df.head())


```

```python
print(df.columns)  # Print column names
print(df.head())   # Print the first few rows to inspect data
```

```python
# Check the data types
print(df.dtypes)

# Check for missing or non-numeric values
print(df['Data.Carbohydrate'].head())  # Print the first few entries in the calories column

# Convert the 'calories' column to numeric, setting errors='coerce' will convert invalid parsing to NaN
df['calories'] = pd.to_numeric(df['Data.Carbohydrate'], errors='coerce')

# Check for NaN values after conversion
print(df['Data.Carbohydrate'].isna().sum())

# Drop rows with NaN values in 'calories' column if any
df = df.dropna(subset=['calories'])

# Ensure 'calories' is in float format
df['calories'] = df['calories'].astype(float)

```

```python
def get_calorie_content(food_item, df):
    # Convert food item to lowercase for case-insensitive matching
    food_item = food_item.lower()

    # Search for the food item in the DataFrame
    row = df[df['Category'].str.lower() == food_item]

    if not row.empty:
        # Return the calorie content
        return row['Data.Carbohydrate'].values[0]
    else:
        # Return a message if the food item is not found
        return "Food item not found in the database."

# Example usage
food = 'milk'
calories = get_calorie_content(food, df)
print(f"The calorie content of {food} is {calories} calories.")
```
